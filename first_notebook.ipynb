```python
%load_ext lab_black
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
colums = [
    "fLength",
    "fWidth",
    "fSize",
    "fConc",
    "fConc1",
    "fAsym",
    "fM3Long",
    "fM3Trans",
    "fAlpha",
    "fDist",
    "class",
]
df = pd.read_csv(
    "ML_dataset/magic+gamma+telescope/magic04.data",
    names=colums,
)
df["class"] = (df["class"] == "g").astype(int)
df
```

```python
fig, axs = plt.subplots(5, 2, figsize=(8, 14))
axs = axs.reshape(-1)
i = 0
for label in df.head(0):
    if label != "class":
        axs[i].hist(df[df["class"] == 1][label], color="b", label="gamma", alpha=0.8)
        axs[i].hist(df[df["class"] == 0][label], color="r", label="hadron", alpha=0.8)
        axs[i].legend()
        axs[i].set_title(label)
        i += 1
plt.tight_layout()
```
